```python
# seperate units depeding on "uM, nM, pM"

for i in range(len(kdAvaCandidate)):
    if 'uM' in kdAvaCandidate['Affinity'][i]:
        kdAvaCandidate['Unit'][i] = 'uM'
        kdAvaCandidate['Affinity'][i] = kdAvaCandidate['Affinity'][i].split('uM')[0]
    elif 'nM' in kdAvaCandidate['Affinity'][i]:
        kdAvaCandidate['Unit'][i] = 'nM'
        kdAvaCandidate['Affinity'][i] = kdAvaCandidate['Affinity'][i].split('nM')[0]
    else:
        kdAvaCandidate['Unit'][i] = 'pM'
        kdAvaCandidate['Affinity'][i] = kdAvaCandidate['Affinity'][i].split('pM')[0]
```

```python
# make *.fasta file

seq_dir = '/mnt/ailon/pdb_test/1110/data/pdbkinase/seq'
for i in range(len(sample)):
    try:
        pdbid = sample['PDBId'][i]
        with open(os.path.join(seq_dir, pdbid + '.fasta'), 'w') as f:
            f.writelines('>'+ sample['PDBId'][i] + '\n')
            f.writelines(sample['Sequence'][i] + '\n')
    except:
        #print(pdbid)
        pass
```

```python
proteins = {}
for i in range(len(sample)):
    proteins[sample['PDBId'][i]] = sample['Sequence'][i]
```
