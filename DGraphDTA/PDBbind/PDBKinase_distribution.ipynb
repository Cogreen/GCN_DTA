# Sample PDBbind data Distribution

### Couting Ligands' Length : sample data(100)

```python
import json

ligands = json.load(open('ligands_can.txt'))
```

```python
ligandLen = []
for i in ligands.keys():
    ligandLen.append(len(ligands[i]))
```

```python
# ligandLen distribution
import matplotlib.pyplot as plt
import seaborn as sns

#atest = sns.load_dataset('atest')
sns.set()
df = ligandLen
sns_plot = sns.distplot(df, hist_kws = {'edgecolor': 'lightpink'}, color='palevioletred')
#plt.xlabel("Length")
#plt.title("Density of Ligand Length")
sns_plot.set(title="Density of Ligand Length", xlabel='Length')
#fig = sns_plot.get_figure()
sns_plot.figure.savefig('ligand_pdb_len.png')
```

### Couting Proteins' Length: sample data(100)

```python
import json

proteins = json.load(open('proteins.txt'))
```

```python
proteinLen = []
for i in proteins.keys():
    proteinLen.append(len(proteins[i]))
```

```python
# proteinLen distribution
import matplotlib.pyplot as plt
import seaborn as sns

#atest = sns.load_dataset('atest')
sns.set()
df = proteinLen
sns_plot = sns.distplot(df, hist_kws = {'edgecolor': 'lightsteelblue'}, color='steelblue')
#plt.xlabel("Length")
#plt.title("Density of Ligand Length")
sns_plot.set(title="Density of Protein Length", xlabel='Length')
#fig = sns_plot.get_figure()
sns_plot.figure.savefig('protein_pdb_len.png')
```
